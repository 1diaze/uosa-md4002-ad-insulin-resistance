```r
library(ggplot2)
library(dplyr)
library(tidyr)
library(gtsummary)
library(ggeffects)
```

```r
tbl1 <- read.csv("028_ADNI Merge Cohort V4")
```

```r
tbl2 <- tbl1[, colnames(tbl1) %in% c("RID", "ABETA", "TAU", "PTAU")]
```

```r
# Converting ABETA, TAU, and PTAU to numeric
tbl2 <- tbl2 |>
    mutate(ABETA == as.numeric(ABETA), na.rm = TRUE) |>
    mutate(TAU = as.numeric(TAU), na.rm = TRUE) |>
    mutate(PTAU = as.numeric(PTAU), na.rm = TRUE
```

```r
# Finding participants with PTAU, TAU, or ABETA Values
tbl3 <- tbl2 |>
    group_by(RID) |>
    mutate(ABETA_sum = sum(ABETA, na.rm = TRUE)) |>
    mutate(TAU_sum = sum(TAU, na.rm = TRUE)) |>
    mutate(PTAU_sum = sum(PTAU, na.rm = TRUE))
tbl4 <- tbl3[, colnames(tbl3) %in% c("RID", "ABETA_sum", "TAU_sum", "PTAU_sum")]
tbl4 <- tbl4 |>
    distinct(RID, PTAU_sum, TAU_sum, ABETA_sum)
tbl4
```

```r
# Taking only participants with CSF datapoints for each of ABETA, TAU, and PTAU
tbl5 <- tbl4 |>
    subset(PTAU_sum > 0) |>
    subset(TAU_sum > 0) |>
    subset(ABETA_sum > 0)
nrow(tbl5)
# 907 Participants with biomarker data for all datapoints
```

```r
# Merging with cohort by RID
tbl6 <- merge(tbl5, tbl1, by = "RID")
```

```r
# Saving as updated cohort w/ medhist + IR + have ALL biomarkers measured at some point
write.csv(tbl6, "030_ADNI Merge Cohort V5")
```

```r
tbl7 <- tbl4 |>
  filter(PTAU_sum > 0 | TAU_sum > 0 | ABETA_sum > 0)
nrow(tbl7)
# 1051 Participants with at least one piece of biomarker data
```

```r
tbl8 <- merge(tbl7, tbl1, by = "RID")
```

```r
# Saving as updated cohort w/ medhist + IR + have at least 1 biomarker measured at some point
write.csv(tbl8, "031_ADNI Merge Cohort V6")
```
