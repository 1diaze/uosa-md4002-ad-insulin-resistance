```r
library(tidyverse)
library(dplyr)
```

```r
phaseI <- read.csv("050_ADNI_Merge_Cohort_V3")
phaseI_bl <- phaseI |>
    subset(VISCODEnum == 0)

phaseII <- read.csv("063_Phase_II_Cohort")
phaseII_bl <- phaseII |>
    subset(VISCODEnum == 0)
```

### NUMERIC VARIABLES

```r
#AGE
wilcox.test(phaseI_bl$AGE, phaseII_bl$AGE)
```

```r
#EDUCATION
wilcox.test(phaseI_bl$PTEDUCAT, phaseII_bl$PTEDUCAT)
```

```r
#BMI
wilcox.test(phaseI_bl$BMI, phaseII_bl$BMI)
```

```r
#MMSE
wilcox.test(phaseI_bl$MMSE_bl, phaseII_bl$MMSE_bl)
```

```r
#Glucose
wilcox.test(phaseI_bl$GLUCOSE, phaseII_bl$GLUCOSE)
```

```r
#TG
wilcox.test(phaseI_bl$TOTAL_TG, phaseII_bl$TOTAL_TG)
```

```r
#HDL-C
wilcox.test(phaseI_bl$HDL_C, phaseII_bl$HDL_C)
```

### CATEGORICAL VARIABLES

```r
phaseI_bl$Group <- "PhaseI"
phaseII_bl$Group <- "PhaseII"

combined <- bind_rows(phaseI_bl, phaseII_bl)
```

```r
#SEX
fisher.test(table(combined$PTGENDER, combined$Group))
```

```r
#Cognitive Status
fisher.test(table(combined$DIAGNOSIS_bl, combined$Group))
```

```r
#Hypertension
fisher.test(table(combined$HYPERTENSION, combined$Group))
```

```r
#Diabetes
fisher.test(table(combined$DIABETES, combined$Group))
```

```r
#Smoking
fisher.test(table(combined$SMOKING, combined$Group))
```

```r
#CVD
fisher.test(table(combined$CVD, combined$Group))
```

```r
#APOE4
fisher.test(table(combined$APOE4, combined$Group))
```
