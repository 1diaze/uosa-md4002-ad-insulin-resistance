```r
library(tidyverse)
library(lme4)
library(lmerTest)
library(broom)
library(broom.mixed)
library(ggeffects)
library(emmeans)
library(performance)
library(patchwork)
library(cowplot)
library(modelsummary)
library(datawizard)
```

```r
df1 <- data.frame(IR_Metric=c('TYG', 'TYG-BMI', 'METS-IR', 'TG/HDL', 'GSK3A', 'GSK3B', 'Insulin'), 
                             AB=c('23.69', '66.76', '72.33', '27.50', '-34.94', '-55.73', '40.13'), 
                             AB_CI_LL=c('0.73', '-12.79', '4.21', '3.33', '-71.27', '-92.98', '0.88'),
                             AB_CI_UL=c('46.67', '146.30', '140.47', '51.68', '1.40', '-18.42', '79.37'),
                             TAU=c('-5.34', '-10.56', '-16.44', '-7.15', '-16.32', '-41.50', '15.95'),
                             TAU_CI_LL=c('-13.81', '-39.13', '-40.95', '-16.06', '-30.44', '-55.46', '0.63'),
                             TAU_CI_UL=c('3.15', '18.07', '8.13', '1.78', '-2.21', '-27.54', '31.25'),
                             PTAU=c('-0.75', '-1.68', '-2.19', '-0.95', '-1.70', '-4.49', '1.57'),
                             PTAU_CI_LL=c('-1.67', '-4.82', '-4.88', '-1.93', '-3.24', '-6.01', '-0.09'),
                             PTAU_CI_UL=c('0.18', '1.47', '0.51', '0.02', '-0.17', '-2.97', '3.24')
                            )
df1
```

```r
df1$AB <- as.numeric(df1$AB)
df1$AB_CI_LL <- as.numeric(df1$AB_CI_LL)
df1$AB_CI_UL <- as.numeric(df1$AB_CI_UL)
df1$TAU <- as.numeric(df1$TAU)
df1$TAU_CI_LL <- as.numeric(df1$TAU_CI_LL)
df1$TAU_CI_UL <- as.numeric(df1$TAU_CI_UL)
df1$PTAU <- as.numeric(df1$PTAU)
df1$PTAU_CI_LL <- as.numeric(df1$PTAU_CI_LL)
df1$PTAU_CI_UL <- as.numeric(df1$PTAU_CI_UL)
df1 <- df1 |>
    mutate(IR_Metric = factor(IR_Metric, levels = c("Insulin", "GSK3B", "GSK3A", "TG/HDL", "METS-IR", "TYG-BMI", "TYG")))
```

```r
ab <- df1[, colnames(df1) %in% c('IR_Metric', 'AB', 'AB_CI_LL', 'AB_CI_UL')]
tau <- df1[, colnames(df1) %in% c('IR_Metric', 'TAU', 'TAU_CI_LL', 'TAU_CI_UL')]
ptau <- df1[, colnames(df1) %in% c('IR_Metric', 'PTAU', 'PTAU_CI_LL', 'PTAU_CI_UL')]
```

### ABETA

```r
abplot <- ab |>
  ggplot(aes(y = IR_Metric)) + 
  theme_classic()
abplot
```

```r
abplot <- abplot +
  geom_point(aes(x=AB), shape=15, size=3) +
  geom_linerange(aes(xmin=AB_CI_LL, xmax=AB_CI_UL)) 
abplot
```

```r
abplot <- abplot +
  geom_vline(xintercept = 0, linetype="dashed") +
  labs(x="Effect on AB", y="IR Metric")
abplot
```

### TAU

```r
tauplot <- tau |>
  ggplot(aes(y = IR_Metric)) + 
  theme_classic()
tauplot
```

```r
tauplot <- tauplot +
    geom_point(aes(x=TAU), shape=15, size=3) +
    geom_linerange(aes(xmin=TAU_CI_LL, xmax=TAU_CI_UL)) +
    geom_vline(xintercept = 0, linetype="dashed") +
    labs(x="Effect on TAU", y="IR Metric")
tauplot
```

### PTAU

```r
ptauplot <- ptau |>
  ggplot(aes(y = IR_Metric)) + 
  theme_classic()
ptauplot
```

```r
ptauplot <- ptauplot +
    geom_point(aes(x=PTAU), shape=15, size=3) +
    geom_linerange(aes(xmin=PTAU_CI_LL, xmax=PTAU_CI_UL)) +
    geom_vline(xintercept = 0, linetype="dashed") +
    labs(x="Effect on PTAU", y="IR Metric")
tauplot
```

```r
ggsave("071_AB_Forest_Plot.png", abplot)
ggsave("072_TAU_Forest_Plot.png", tauplot)
ggsave("073_PTAU_Forest_Plot.png", ptauplot)
```
